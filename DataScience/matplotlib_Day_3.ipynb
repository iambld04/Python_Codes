```python
time_of_day = ['Dawn', 'Morning','Brunch','Noon','Afternoon','Dusk','Evening','Midnight']
sales_day1 = [50,78,42,20,57,72,102,45]
sales_day2 = [45,85,33,24,45,70,85,40]
sales_day3 = [40,80,25,15,38,66,82,56]
```

```python
plt.figure(figsize=(12,6))
plt.plot(time_of_day,sales_day1,color='red',label='Day 1',linestyle='--',marker='*')
plt.plot(time_of_day,sales_day2,color='blue',label='Day 2',linestyle=':',marker='*')
plt.plot(time_of_day,sales_day3,color='green',label='Day 3',linestyle='-.',marker='*')
plt.legend(loc='upper left')
plt.xticks(rotation=40)
plt.yticks(np.arange(10,110,10))
plt.xlabel('Time of Day',fontsize=11)
plt.ylabel('Number of cups of tea sold')
plt.title('Sales trend for the tea over the 3 days')
plt.grid(linestyle='--',axis='y')
plt.show()
```

```python
clr = ['b','g','r','orange','pink','gold','lightyellow','purple']
plt.figure(figsize=(12,5))
plt.bar(time_of_day,sales_day1,width=.7,edgecolor='b',linestyle='--',fill=True,color=clr)
plt.show()
```
