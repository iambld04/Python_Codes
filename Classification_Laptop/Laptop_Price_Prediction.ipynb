```python
def remove_cpu(text):
    if 'intel' in text.lower():
        if 'core' not in text.lower():
            return " ".join(text.split(' ')[0:2])
        else:
            return text
    elif 'amd' in text.lower():
        if 'amd' and 'ryzen' in text.lower():
            return " ".join(text.split(' ')[0:2])+' Series'
        elif text[4]=='A':
            return 'AMD A Series'
        elif text[4]=='E':
            return 'AMD E Series'
        elif text[4]=='F':
            return 'AMD F Series'
```

```python
df_data['Cpu_mid'] = df_data['Cpu_mid'].apply(remove_cpu)
```

```python
def remove_op(text):
    if 'windows' in text.lower():
        if '7' in text.lower():
            return 'Windows 7'
        else:
            return 'Windows 10'
    elif 'mac' in text.lower():
        return 'MacOS'
    else:
        return "No Os/Linux/Others"
```

```python
df_data['Memory'] = df_data['Memory'].astype(str).replace('\.0',"",regex=True)
```

```python
df_data['Memory'].value_counts()
```

```python
df_data['Memory'] = df_data['Memory'].str.replace("GB","")
df_data['Memory'] = df_data['Memory'].str.replace("TB","000")
```

```python
df_data['Memory'].value_counts()
```

```python
new = df_data['Memory'].str.split('+',expand=True)  #expand=True makes data to be in DataFrame and if false it will be in series
df_data['first'] = new[0]
df_data['first'] = df_data['first'].str.strip()
df_data['second'] = new[1]
```

```python
df_data['Layer1HDD'] = df_data['first'].apply(lambda x: 1 if 'HDD' in x else 0)
df_data['Layer1SSD'] = df_data['first'].apply(lambda x: 1 if 'SSD' in x else 0)
df_data['Layer1Hybrid'] = df_data['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df_data['Layer1Flash'] = df_data['first'].apply(lambda x: 1 if 'Flash' in x else 0)
```

```python
df_data['first'] = df_data['first'].str.replace(r'\D',"",regex=True)
```

```python
df_data['second'].fillna('0',inplace=True)
```

```python
df_data['Layer2HDD'] = df_data['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df_data['Layer2SSD'] = df_data['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df_data['Layer2Hybrid'] = df_data['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df_data['Layer2Flash'] = df_data['second'].apply(lambda x: 1 if 'Flash' in x else 0)
```

```python
df_data['second'] = df_data['second'].str.replace(r'\D',"",regex=True)
```

```python
df_data['first'] = df_data['first'].astype('int')
df_data['second'] = df_data['second'].astype('int')
```

```python
df_data.sample()
```

```python
df_data.info()
```

```python
df_data['HDD'] = df_data['first']*df_data['Layer1HDD'] + df_data['second']*df_data['Layer2HDD']
df_data['SSD'] = df_data['first']*df_data['Layer1SSD'] + df_data['second']*df_data['Layer2SSD']
df_data['Hybrid'] = df_data['first']*df_data['Layer1Hybrid'] + df_data['second']*df_data['Layer2Hybrid']
df_data['Flash_Storage'] = df_data['first']*df_data['Layer1Flash'] + df_data['second']*df_data['Layer2Flash']
```

```python
df_data.drop(columns=['first','second','Layer1HDD','Layer1SSD','Layer1Hybrid','Layer1Flash','Layer2HDD','Layer2SSD','Layer2Hybrid','Layer2Flash'],inplace=True)
```
