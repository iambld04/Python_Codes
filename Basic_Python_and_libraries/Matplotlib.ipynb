# Theres two type of data
1.  Categorical Data
2.  Measurable Data

So we need to think for which data what type of graph should be plotted

# If the X-axis is categorical and the Y-axis is measurable then we need to use bar graph

# If the X-axis date or time and the y-axis measurable then we need to use line graph

# If the X-axis measurable and the y-axis is also measurable we need to use scatter plot

# If the x-axis is categorical and the y-axis is measurable we can use pie chart but we need to use bar chart if there are more than 5-7 columns 

# Histogram x-axis can be anything and y-axis is count/density/frequency

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a1 = [1,2,3,4,5]
a2 = [6,7,8,9,10]
```

```python
plt.scatter(a1,a2)
```

```python
plt.plot(a1,a2)
```

```python
plt.barh(a1,a2)
```

```python
plt.hist(a1)
```

```python
time_of_day = ['Dawn','Morning','Brunch','Noon','Afternoon','Dusk','Evening','Midnight']
salary_day_1 = [50,78,42,20,57,72,102,45]
salary_day_2 = [45,85,33,24,45,70,85,40]
salary_day_3 = [40,80,25,15,38,66,82,56]
```

```python
plt.figure(figsize=(12,5))
plt.plot(time_of_day,salary_day_1,color='red',linestyle='--',label='Day1',marker='*')
plt.plot(time_of_day,salary_day_2,color='b',marker='*',linestyle=':',label='Day2')
plt.plot(time_of_day,salary_day_3,color='g',marker='*',linestyle='-.',label='Day2')
plt.legend(loc='upper left')
plt.xticks(rotation=40)
plt.yticks(np.arange(10,111,10))
plt.xlabel('Time of Day',fontsize=15)
plt.ylabel('Number of tea cups sold')
plt.title('sales trend for tea over the past 3 days')
plt.grid(linestyle='--',axis='y')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.bar(time_of_day,salary_day_1,width=0.4,edgecolor='black',fill=True,color=['orange','yellow','green'])
plt.show()
```

```python
plt.pie(salary_day_1,labels=time_of_day,autopct='%0.2f',startangle=90)
plt.show()
```
